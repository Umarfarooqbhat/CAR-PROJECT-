```python
from sklearn.impute import SimpleImputer
```

```python
impute=SimpleImputer(strategy = 'most_frequent')
data[['make', 'model', 'trim', 'body', 'transmission', 'state', 'color',
       'interior', 'seller', 'saledate']]=impute.fit_transform(data[['make', 'model', 'trim', 'body', 'transmission', 'state', 'color',
       'interior', 'seller', 'saledate']])
```

```python
impute=SimpleImputer(strategy = 'median')
data[['condition', 'odometer', 'mmr', 'sellingprice']]=impute.fit_transform(data[['condition', 'odometer', 'mmr', 'sellingprice']])
```

```python
data.isna().sum().sum()/data.shape[0]
```

# 4 handling outliers

```python
sns.boxplot(data[['condition', 'odometer', 'mmr', 'sellingprice']])
```

```python
q1=data[['condition', 'odometer', 'mmr', 'sellingprice']].quantile(0.25)
q3=data[['condition', 'odometer', 'mmr', 'sellingprice']].quantile(0.75)
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
data=data[(data['condition']>lb['condition']) & (data['condition']<ub['condition']) &
(data['odometer']>lb['odometer']) & (data['odometer']<ub['odometer'])&
(data['mmr']>lb['mmr']) & (data['mmr']<ub['mmr'])&
(data['sellingprice']>lb['sellingprice']) & (data['sellingprice']<ub['sellingprice'])]
```

```python
sns.boxplot(data[['condition', 'odometer', 'mmr', 'sellingprice']])
```

```python
data.reset_index(drop=True,inplace=True)
```
